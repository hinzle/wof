## sns.relplot (figure lvl function)

```python
tips = sns.load_dataset('tips')
```

```python
tips.dtypes
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# scatterplot in matplotlib

plt.scatter(tips.total_bill, tips.tip)
plt.title('Tips by Bill')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
```

### col

```python
sns.relplot(x='total_bill', y='tip', col='time', data=tips)
```

### hue

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', style='smoker', data=tips)
```

```python
# Style argument: extra dimension
type(sns.relplot(
    data = tips,
    x = 'total_bill',
    y = 'tip',
    style = 'time',
    hue='day'))
```

```python
sns.relplot(x='total_bill', y='tip', col='time', hue='size', style='smoker', data=tips)
```

### kind

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind='line')
```

```python
# lineplot for discrete values. The lineplot will use mean value at each size to draw a line 
# shaded region is 95% CI. Can be turned off with 'ci = None'

sns.relplot(data = tips, x = 'size', y = 'tip', kind = 'line', ci=None)
```

### small multiples

```python
#  FacetGrid with 'col'
sns.relplot(data = tips, x = 'total_bill', y = 'tip',
            hue = 'time', col = 'day')
```

```python
# Facetgrid with lineplot

sns.relplot(data = tips, x = 'size', y = 'tip', kind = 'line', col = 'day')
```

## sns.displot

```python
sns.displot(data = tips, x = 'total_bill', kde = True, rug = True)

# kde=True approximates the distribution for continuous values
```

```python
sns.displot(data = tips, x = 'total_bill', hue = 'sex', multiple = 'stack',
            palette = ['r', 'b'], kind = 'kde', alpha = 0.4)
```

# Categorical

## sns.boxplot

```python
fig, ax = plt.subplots(2,2, figsize = (10,8))

sns.boxplot(data = tips, y = 'tip', ax = ax[0,0])

sns.boxplot(data = tips, x = 'tip', ax = ax[0,1])

sns.boxplot(data = tips, y = 'tip', x = 'size', ax = ax[1,0])

sns.boxplot(data = tips, x = 'tip', y = 'time', hue = 'smoker', ax = ax[1,1])
```

```python
# old way

plt.figure(figsize=(12, 10))
plt.suptitle('Boxplots with Seaborn')

plt.subplot(221)
sns.boxplot(data=tips, y='tip')
plt.title('A "typical" boxplot')

plt.subplot(222)
sns.boxplot(data=tips, x='tip')
plt.title('Rotated 90$\degree$')

plt.subplot(223)
sns.boxplot(data=tips, y='tip', x='size')
plt.title('Providing an x and y value groups the data')

plt.subplot(224)
sns.boxplot(data=tips, y='time', x='tip', hue='smoker')
plt.title('You can also use other kwargs, like hue')

plt.subplots_adjust(hspace=0.4)
```

```python
from env import get_db_url
```

```python
url=get_db_url('chipotle')
```

```python
chipotle=pd.read_sql(
	'''
	SELECT * FROM orders
	'''
	,url
)
```

```python
chipotle.sample(10)
```

```python
chipotle['item_price'] = chipotle.item_price.str.replace('$', '').astype('float')
```

```python
chipotle.sample(10)
```

```python
chipotle['chip_otal']=chipotle['quantity']*chipotle['item_price']
chipsss=chipotle.groupby('item_name').chip_otal.sum().nlargest(4)
chipsss
```

```python
dipsss=pd.DataFrame({'item':chipsss.index,'revenue':chipsss.values})
dipsss.head()
```

```python
sns.barplot(data=dipsss,x='revenue',y='item')
```
