```python
n1 = list(np.linspace(-10, 10, num=10000))
y1 = [(0 * n + 1) for n in n1]

n2 = list(range(1, 100))
y2 = [math.log(n) for n in n2]

n3 = list(np.linspace(-10, 10, num=10000))
y3 = [n for n in n3]

n4 = list(range(1, 100))
y4 = [n * math.log(n) for n in n4]

n5 = list(np.linspace(-10, 10, num=10000))
y5 = [n**2 for n in n5]

n6 = list(range(11))
y6 = [math.factorial(n) for n in n6]

n7 = list(np.linspace(-10, 10, num=10000))
y7 = [2**n for n in n7]
```

```python
plt.figure(figsize=(16, 9))

plt.plot(n1, y1, label='$O(1)$')
plt.plot(n2, y2, label='$O(log n)$')
plt.plot(n3, y3, label='$O(n)$')
plt.plot(n4, y4, label='$O(n log n)$')
plt.plot(n5, y5, label='$O(n^2)$')
plt.plot(n6, y6, label='$O(n!)$')
plt.plot(n7, y7, label='$O(2^n)$')

plt.xlim(-10, 10)
plt.ylim(-1, 10)

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')

plt.legend(loc='upper left')
```
