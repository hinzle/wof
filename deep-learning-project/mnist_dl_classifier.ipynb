```python
# tensorflow imports
from tensorflow import keras
from keras import models, layers

from keras.datasets import mnist
```

```python
# acquire:
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# what does this data set look like?
print(test_images.shape)
print(train_images.shape)
```

```python
# reshape our information to make it conducive to the shape of our ANN,
# min-max scaling to establish values between 0 and 1 for greyscale color channel
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
  
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
# model
# established what model I want to use
network = models.Sequential()
```

```python
# add my layers, specifying what activations and shapes I want to utilize
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, activation='softmax'))
```

```python
# compile our ANN, cementing structure and defining our optimizer, loss and metrics
network.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python

# encode our labels appropriately
train_labels = keras.utils.to_categorical(train_labels)
test_labels = keras.utils.to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=20, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print(f'accuracy of network on test set: {test_acc}')
```
